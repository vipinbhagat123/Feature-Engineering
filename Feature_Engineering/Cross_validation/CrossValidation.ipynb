```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("/content/placement.csv")
dataset
```

```python
x=[['cgpa']]
y=['package']
```

```python
new_data=dataset.head(10)
```

```python
new_data
```

```python
x_new=new_data.iloc[:,:-1]
y_new=new_data['package']
```

```python
from sklearn.model_selection import LeaveOneOut, LeavePOut, KFold, StratifiedGroupKFold
```

```python
# LeaveOneOut Cross Validation Technique
lo=LeaveOneOut()
for train, test in lo.split(x_new, y_new):
  print(train, test)
```

```python
# LeavePOut Cross Validation Technique
# We can customize the p
lo=LeavePOut(p=2)
for train, test in lo.split(x_new, y_new):
  print(train, test)
```

```python
# KFold Cross Validation Technique
# We can customize the n_splits
lo=KFold(n_splits=5)
for train, test in lo.split(x_new, y_new):
  print(train, test)
```
