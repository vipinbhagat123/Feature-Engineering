```python
import pandas as pd
```

```python
da=pd.DataFrame({"size":['s', 'm', 'l', 'xl', 's', 'm', 'l', 's', 's', 'l', 'xl', 'm']})
da
```

```python
ord_da=[["s", "m", "l", "xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ord_da)
oe.fit(da[["size"]])
```

```python
da['en_size']=oe.transform(da[["size"]])
```

```python
da
```

```python
ord_data1={"s":0, "m":1, "l":2, "xl":3}
```

```python
da['size_en_map']=da["size"].map(ord_data1)
```
