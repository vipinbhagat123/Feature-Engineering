```python
import pandas as pd
```

```python
da=pd.read_csv("/content/loan_data_set.csv")
da
```

```python
da.isnull().sum()
```

```python
da['Gender'].fillna(da['Gender'].mode()[0], inplace=True)
```

```python
da['Married'].fillna(da['Married'].mode()[0], inplace=True)
```

```python
da.isnull().sum()
```

```python
en=da[['Gender', 'Married']]
en
```

```python
pd.get_dummies(en)
```

```python
 pd.get_dummies(en).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
ar=ohe.fit_transform(en).toarray()
```

```python
pd.DataFrame(ar, columns=[" Gender_Female", "Gender_Male", "Married_NO", "Married_Yes"])
```

```python
ohe=OneHotEncoder(drop="first")
ar=ohe.fit_transform(en).toarray()
ar
```

```python
pd.DataFrame(ar, columns=['Gender_', 'Married_'])
```
