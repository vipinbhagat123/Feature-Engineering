```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
da=pd.read_csv("/content/loan_data_set.csv")
da.head(3)
```

```python
da.isnull().sum()
```

```python
da.describe()
```

```python
sns.boxplot(x="CoapplicantIncome", data=da)
```

```python
sns.distplot(da["CoapplicantIncome"])
```

```python
min_range=da["CoapplicantIncome"].mean()-(3*da["CoapplicantIncome"].std())
max_range=da["CoapplicantIncome"].mean()+(3*da["CoapplicantIncome"].std())
```

```python
min_range, max_range
```

```python
new_da=da[da["CoapplicantIncome"]<=max_range]
```

```python
sns.boxplot(x="CoapplicantIncome", data=new_da)
```

```python
#Using Z_score
z_score=(da["CoapplicantIncome"]-da["CoapplicantIncome"].mean()/(da["CoapplicantIncome"].std()))
```

```python
z_score
```

```python
z_score>3
```

```python
da["z_score"]=z_score
```
