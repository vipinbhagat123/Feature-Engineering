# **The interquartile range (IQR) method is a way to identify outliers in a set of data.**

```python
da.shape
```

```python
q1=da["CoapplicantIncome"].quantile(0.25)
q3=da["CoapplicantIncome"].quantile(0.75)
q1,q3
```

```python
IQR=q3-q1
```

```python
min_range=q1-(1.5*IQR)
max_range=q3+(1.5*IQR)
min_range, max_range
```

```python
new_da=da[da["CoapplicantIncome"]<=max_range]
```

```python
new_da.shape
```

```python
sns.boxplot(x="CoapplicantIncome", data=new_da)
plt.show()
```
