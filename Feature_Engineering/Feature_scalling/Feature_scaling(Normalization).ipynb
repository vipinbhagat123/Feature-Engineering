```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
da=pd.read_csv("/content/loan_data_set.csv")
da
```

```python
da.head(3)
```

```python
da.isnull().sum()
```

```python
da["ApplicantIncome"].fillna(da["ApplicantIncome"].mean(), inplace=True)
```

```python
sns.displot(da["ApplicantIncome"])
plt.show()
```

```python
da.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(da[["ApplicantIncome"]])
```

```python
ss.transform(da[["ApplicantIncome"]])
```

```python
da["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(da[["ApplicantIncome"]]),columns=['x'])
```

```python
da.head(3)
```

```python
da.describe()
```

```python
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(da["ApplicantIncome_ss"])
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(da["ApplicantIncome"])
plt.show()
```
