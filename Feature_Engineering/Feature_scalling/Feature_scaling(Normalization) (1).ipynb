## **Normalization: It is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max Scaling.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
da=pd.read_csv("/content/loan_data_set.csv")
da
```

```python
da.head(3)
```

```python
da.isnull().sum()
```

```python
da.describe()
```

```python
sns.distplot(da["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(da[["CoapplicantIncome"]])
```

```python
da["CoapplicantIncome_min_max"]=ms.transform(da[["CoapplicantIncome"]])
```

```python
da.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(da["CoapplicantIncome"])
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(da["CoapplicantIncome_min_max"])
plt.show()
```
